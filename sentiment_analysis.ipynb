```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv("/content/IMDB Dataset.csv.zip",nrows=35000)
df.head()
```

```python
df.shape
```

```python
df.info()
df.isnull().sum()
```

```python
df["sentiment"]=df["sentiment"].astype('category').cat.codes
df["sentiment"]=df["sentiment"].astype(int)
df.head()
```

```python
df["review"][0]
```

```python
import re
```

```python
def clear_html_tags(text):
  #pattern=re.compile(r'<br\s*/?>')
  pattern = re.compile(r'<br\s*/?>|> >')
  rt=re.sub(pattern," ",text)
  return rt
```

```python
df["review"]=df["review"].apply(clear_html_tags)
df.head()
```

```python
df["review"][0]
```

```python
def lowercase(text):
  return text.lower()
```

```python
df["review"]=df["review"].apply(lowercase)
df.head()
```

```python
def remove_special_char(text):
  pattern=re.compile(r'[^a-zA-z0-9\s]')
  rt=re.sub(pattern,"",text)
  return rt
```

```python
df["review"]=df["review"].apply(remove_special_char)
df.head()
```

```python
df["review"][0]
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python

def remove_stopwords(text):
   x=[]
   stopwrd=set(stopwords.words("english"))
   for i in text.split():
        if i not in stopwrd:
           x.append(i)
   y=x[:]
   x.clear()
   return " ".join(y)


```

```python
df["review"]=df["review"].apply(remove_stopwords)
df.head()
```

```python
df["review"][0]
```

```python
from nltk.stem.snowball import SnowballStemmer
ps=SnowballStemmer("english")
```

```python
def stemming(text):
    x=[]
    for i in text.split():
        x.append(ps.stem(i))
    y=x[:]
    x.clear()
    return " ".join(y)
```

```python
df["review"]=df["review"].apply(stemming)
df.head()
```

```python
df["review"][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=25000)
```

```python
x=df["review"]
y=df.iloc[:,[-1]]
```

```python
x=cv.fit_transform(x)
```

```python
x=x.toarray()
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
from sklearn.naive_bayes import MultinomialNB,GaussianNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
cf1=MultinomialNB()
cf2=GaussianNB()
cf3=BernoulliNB()
```

```python
cf1.fit(x_train,y_train)
```

```python
y_pred=cf1.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
cf3.fit(x_train,y_train)
```

```python
y_pred2=cf3.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred2))
```
